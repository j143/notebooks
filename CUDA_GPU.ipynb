```
# Apache Maven Installation script

# Run and print a shell command.
import os

def run(command):
  print('>> {}'.format(command))
  !{command}
  print('')

# Download the maven source.
maven_version = '3.9.0'
maven_path = f"/opt/apache-maven-{maven_version}"

if not os.path.exists(maven_path):
  run(f"wget -q -nc -O apache-maven.zip https://dlcdn.apache.org/maven/maven-3/{maven_version}/binaries/apache-maven-{maven_version}-bin.zip")
  run('unzip -q -d /opt apache-maven.zip')
  run('rm -f apache-maven.zip')

# Let's choose the absolute path instead of $PATH environment variable.
def maven(args):
  run(f"{maven_path}/bin/mvn {args}")

maven('-v')
```
